# **Proses Training dan Testing**

**Membagi 150 Data menjadi data untuk training dan data untuk testing**<Br>
**Dicoba membagi data training (75%) dan data testing (25%)**

```python
from sklearn.model_selection import train_test_split
#x = data, y = target
x_train, x_test, y_train, y_test = train_test_split(
    iris_datasets['data'],iris_datasets['target'], test_size = 0.25, random_state = 0)
#random_state = 0 -> kalau di run ulang, isi data yang dibagi tetap sama
```

```python
#jumlah data training
print("x_train shape : {} ".format(x_train.shape))#DATA = ( Sepal Length, sepal widt, petal length, petal widt)
print("y_train shape : {} ".format(y_train.shape))#TARGET = ( Sentosa, versicolor, verginical)
```

```python
#jumlah data training
print("x_test shape : {} ".format(x_test.shape))#DATA = ( Sepal Length, sepal widt, petal length, petal widt)
print("y_test shape : {} ".format(y_test.shape))#TARGET = ( Sentosa, versicolor, verginical)
```

# **PROSES TRAINING**

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 1)
```

```python
knn.fit(x_train, y_train)
```

melakukan prediksi

```python
import numpy as np
x_baru = np.array([[5,2.1, 1, 0.2]])
prediksi = knn.predict(x_baru)
print("HASIL PREDIKSI KNN = {}".format(prediksi))
print("HASIL PREDIKSI NAMA BUNGA = {}".format(iris_datasets['target_names'][prediksi]))
```

```python
import numpy as np
x_baru = np.array([[3,5.1, 5, 4.2]])
prediksi = knn.predict(x_baru)
print("HASIL PREDIKSI KNN = {}".format(prediksi))
print("HASIL PREDIKSI NAMA BUNGA = {}".format(iris_datasets['target_names'][prediksi]))
```

# EVALUASI MODEL

UKUR AKURASI

```python
y_pred = knn.predict(x_test)
print(y_pred)
```

```python
print("HASIL AKURAASI MODEL = {}".format(knn.score(x_test,y_test)))
#hasil akurasi 97%
```
