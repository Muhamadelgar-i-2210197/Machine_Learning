# <b> Principal component Analysis </b> <br>
Apa itu PCA ? <br>

PCa Merupakan metode statistik yang di gunakan untuk <b> Mereduksi Dimensi</b> dengan cara mengubah data kedalam komponen utama (principal coponent) yang memiliki variasi terbesar, PCA diguanakan untuk mengatasi masalah dimensi padaa data yang memiliki banyak fitur, tanpa kehilangan terlalu banyak informasi penting

<b> Manfaat PCA : </b>
1. eduksi Dimensi
2. Visualisasi Data
3. Mengurangi Overfitting ( menghapus fitur kurang penting )
4. Mempercepat Komputasi

**Klasifikasi menggunakan Support Vector Machine(Pakai PCA)**

```python
x = data[['Algebra', 'Calculus1','Calculus2', 'Statistics', 'Probability', 'Measure','Functional_analysis']]
y = data['y']
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
#preprocessing
#Normalisasi DAta ( data di - scaling agar memiliki varians yang seragam atau sama)
Scaler = StandardScaler()
x_scaled = Scaler.fit_transform(x)
```

```python
pca = PCA(n_components = 2)# mengurangi fitur jadi hanya ada 2 fitur
x_pca = pca.fit_transform(x_scaled)
```

```python
x_pca
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_pca, y, test_size= 0.2, random_state= 0)
```

```python
#support Vector classification
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

model_svc = SVC()
model_svc.fit(x_train,y_train)

y_pred = model_svc.predict(x_test)
akurasi = accuracy_score(y_test, y_pred)
print(f'Hsdil Akurasi Menggunakan SVM = {akurasi:.2%}')
```

Daari Hasil pertama Data yang digunakan untuk training hanya menggunakan 2 fitur saja hanayaa memberikan hasil akuraasi sbesar 61.95 saja. Sedaangkan menggunakana Data dengan 5 fitur dan lalu di reduksi menjadi 2 fitur agar informasi dari data ssebelumnya tidak hilang memberikan hasil akurasi sebesaar 66.7% <br>

Kessimpulannya: <br>
Aagar dapat mempertahankan tnpa mengurangi jumlah akurasi yang besar, di sarankan untukmengambil fitur sebanyak baanyaknyaa daan melakukan reduksi menggunakan PCA jika tujuannya untuk Visualisasi 2 dimensi  (2 fitur ) atau 3 dimenssi ( 3 fitur)
