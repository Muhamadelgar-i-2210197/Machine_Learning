# **Library Pandas**

digunakan untuk manipulasi dan analisis data

```python
import pandas as pd

data = {"Nama" : ['Andi',"Mulyono","Budi"],
        "Umur" : [21, 30, 35],
        "Kota" : ["Jakarta", "Solo", "Bogor"]}
data2 = pd.DataFrame(data)
display(data2)
```

```python
display(data2[data2.Umur >= 30])
```

```python
display(data2[data2.Nama == "Mulyono"])
```

```python
display(data2[data2.Nama != "Andi"])
```

```python
display(data2[(data2.Nama == "Mulyono") | (data2.Nama == "Budi")])
```
