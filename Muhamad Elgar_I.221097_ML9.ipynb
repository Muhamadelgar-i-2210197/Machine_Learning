# **Home**

```python
data_home = data[data['Product Category'] == 'Home']
data_home
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

data_home = data[data['Product Category'] == 'Home']

x = data_home[['Product Price','Quantity','Customer Age']] #Features
y = data_home[['Total Purchase Amount']] #target

#Split Data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)

#Training
model = LinearRegression()
model.fit(x_train, y_train)

#Evaluasi
y_pred_train = model.predict(x_train)
y_pred = model.predict(x_test)

# Evaluasi Training
mse_train = mean_squared_error(y_pred_train,y_train)
print(f'Hasil Mean Squared Error Training = {mse_train:.3f}')

#Testing
mse = mean_squared_error(y_pred,y_test)
print(f'Hasil Mean Squared Error Testing = {mse:.3f}')
```

hasil dari metode yang digunakan yaitu linear regresission menunjukkan bahwa hasil tersebut underfitting

```python
import numpy as np
feature_importance = pd.DataFrame({
    'Feature' : x.columns,
    'Importance' : model.coef_[0]
})

feature_importance = feature_importance.sort_values(by = 'Importance', ascending= False)
print (feature_importance)
```

Penjualan total untuk produk kategori Home dipengaruhi oleh Customer Age

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

data_perempuan = data[data['Gender'] == 'Female']
data_perempuan_mean = data_perempuan[data_perempuan['Customer Age'] > data_perempuan['Customer Age'].mean()]

x = data_perempuan_mean[['Product Price', 'Quantity', 'Customer Age']] #Features
y = data_perempuan_mean[['Total Purchase Amount']] #target

#Split Data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)

#Training
model = LinearRegression()
model.fit(x_train, y_train)

#Evaluasi
y_pred = model.predict(x_test)

mse = mean_squared_error(y_pred,y_test)
print(f'Hasil Menan Squared Error = {mse:.3f}')
```

Hasil Mean Squared Error (MSE) = 2069618.737 pada data perempuan dengan usia rata-rata di atas menunjukkan bahwa model mungkin mengalami overfitting

```python
import numpy as np
feature_importance = pd.DataFrame({
    'Feature': x.columns,
    'Importance': model.coef_[0]
})

feature_importance = feature_importance.sort_values(by='Importance',ascending= False)
print(feature_importance)
```


Dari hasil feature importance, Customer Age memiliki pengaruh yang lebih besar pada data laki-laki dengan nilai 5.435304, dibandingkan dengan perempuan yang memiliki nilai 5.119484. Jadi tinggi atau banyak pada laki-laki
